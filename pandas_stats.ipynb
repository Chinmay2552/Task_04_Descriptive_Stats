```python
import pandas as pd

# Load the dataset
df = pd.read_csv('2024_fb_posts_president_scored_anon.csv')

# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include='number').columns.tolist()
non_numeric_cols = df.select_dtypes(exclude='number').columns.tolist()

# Descriptive statistics for numeric columns
print("=== Descriptive Statistics (Numeric Columns) ===")
print(df[numeric_cols].describe())

# Unique and most frequent values for non-numeric columns
print("\n=== Unique and Most Frequent Values (Non-Numeric Columns) ===")
for col in non_numeric_cols:
    print(f"\nColumn: {col}")
    print(f"  Unique Values: {df[col].nunique()}")
    print(f"  Most Frequent:\n{df[col].value_counts().head(1)}")
```

```python


import polars as pl
from collections import Counter

# Load dataset
df = pl.read_csv("/content/2024_fb_posts_president_scored_anon.csv")

# Separate column types
numeric_cols = [col for col, dtype in zip(df.columns, df.dtypes) if dtype in pl.NUMERIC_DTYPES]
non_numeric_cols = [col for col in df.columns if col not in numeric_cols]

# === Descriptive stats for numeric columns ===
print("=== Descriptive Statistics (Numeric Columns) ===")
print(df.select(numeric_cols).describe())

# === Unique and Most Frequent for Non-Numeric Columns ===
print("\n=== Unique and Most Frequent Values (Non-Numeric Columns) ===")
for col in non_numeric_cols:
    col_values = df.select(col).to_series().to_list()
    unique_vals = set(col_values)
    most_common = Counter(col_values).most_common(1)[0]

    print(f"\nColumn: {col}")
    print(f"  Unique Values: {len(unique_vals)}")
    print(f"  Most Frequent: {most_common}")
```
